```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Reading in the data
df = pd.read_csv(r"C:\Users\mrmar\Documents\Data Analyst Bootcamp\Python and Pandas Datasets\movies.csv")


```

```python
df.head()
```

```python
# Data Cleaning
# First, we are checking for missing data.

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types of our columns

df.dtypes
```

```python
pd.set_option('display.max_rows', None)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Let us check to see if there are any outliers.
df.boxplot(column=['gross'])
```

```python
# We will be checking for which variables are most correlated to Gross Revenue.
```

```python
# Scatter plot with Budget vs Gross Revenue


plt.scatter(x = df['budget'], y = df['gross'])
plt.title("Budget vs Gross Revenue")
plt.xlabel('Gross Revenue')
plt.ylabel("Budget of Film")
plt.show()
```

```python
df.head()
```

```python
#Create a regression plot for budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color" : "black"})
```

```python
# Now let us start checking for the correlation between these two variables.

df.corr(numeric_only = True, method = 'pearson') 
```

```python
#We can see that there is a high correlation between the film budget and the gross revenue.

corr_mat = df.corr(numeric_only = True, method = 'pearson') 
sns.heatmap(corr_mat, annot = True)
plt.title("Correlation Matrix of Numeric Variables")
plt.xlabel('Movie Features')
plt.ylabel("Movie Features")
plt.show()
```

```python
# Now let us look at the correlation between the company and the gross revenue


df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
corr_mat2 = df_numerized.corr(numeric_only = True, method = 'pearson') 
sns.heatmap(corr_mat2, annot = True)
plt.title("Correlation Matrix of All Variables")
plt.xlabel('Movie Features')
plt.ylabel("Movie Features")
plt.show()
```

```python
df_numerized.corr()
```

```python
cor_mat = df_numerized.corr()
cor_pairs = cor_mat.unstack()
cor_pairs
```

```python
sorted_pairs = cor_pairs.sort_values()
sorted_pairs
```

```python
high_cor = sorted_pairs[(sorted_pairs)>0.5]
high_cor
```

```python
# Therefore, we found that the budget and the fan votes have the highest correlation to gross revenue. 
# We also found that the company had low correlation to the gross revenue.
```
